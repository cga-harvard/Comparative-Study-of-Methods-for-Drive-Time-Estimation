## Read Data

```python
import pandas as pd
results = pd.read_csv("../data/processed/sample_and_results.csv")
results.head()
```

## Visualize Results

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

results.index = results["Google Maps"]
ax = results.plot(kind='line',
        y=["Bing Maps", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Pro"],
        linewidth=0.3,alpha=0.9)
# make the line width of google maps bigger
results.plot(kind='line',y=["Google Maps"],linewidth=1.5,ax=ax)

# plt.title("Comparison of Route Drive Durations")
plt.xlabel("Google Maps Drive Time (Minute)")
plt.ylabel("Drive Time (Minute)")
# ax.set_xlim([0, 800])
# ax.set_ylim([0, 800])
# set legend line width

legend_lines = [mlines.Line2D([], [], color=c, linewidth=2) for c in plt.rcParams['axes.prop_cycle'].by_key()['color']]
legend_labels = [ "Bing Maps", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Pro","Google Maps"]
plt.legend(legend_lines, legend_labels,frameon=False)


# plt.savefig("../reports/figures/Comparison_of_Route_Drive_Durations.png",dpi=300)
# plt.legend(["Google Map", "Bing Map", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Desktop"])

plt.show()
```

```python
# use plotly to make interactive plot
import plotly.express as px

fig = px.line(results, x="Google Maps", y=["Bing Maps",
                                            "ESRI Routing Service", "OSRM", 
                                            "OSMNx", "ArcGIS Pro","Google Maps"])
fig.update_layout(
    title="Comparison of Route Drive Durations",
    xaxis_title="Google Maps Drive Time (Minute)",
    yaxis_title="Drive Time (Minute)",
    legend_title="",
    font=dict(
        family="Courier New, monospace",
        size=12,
        color="black"
    )
)
fig.update_traces(line=dict(width=1.5))
fig.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

results.index = results["Google Maps"]
ax = results.plot(kind='line',
        y=["Bing Maps", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Pro"],
        linewidth=0.1,alpha=1)
# make the line width of google maps bigger
results.plot(kind='line',y=["Google Maps"],linewidth=1.5,ax=ax)

# plt.title("Comparison of Route Drive Durations")
plt.xlabel("Google Maps Drive Time (Minute)")
plt.ylabel("Drive Time (Minute)")
ax.set_xlim([0, 100])
ax.set_ylim([0, 150])
# set legend line width
# plt.title("Comparison of Route Drive Durations")
plt.xlabel("Google Maps Drive Time (Minute)")
plt.ylabel("Drive Time (Minute)")
legend_lines = [mlines.Line2D([], [], color=c, linewidth=2) for c in plt.rcParams['axes.prop_cycle'].by_key()['color']]
legend_labels = [ "Bing Maps", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Pro","Google Maps"]
plt.legend(legend_lines, legend_labels,frameon=False)

plt.savefig("../reports/figures/Comparison_of_Route_Drive_Durations_with100.pdf")
# plt.legend(["Google Map", "Bing Map", "ESRI Routing Service", "OSRM", "OSMNx", "ArcGIS Desktop"])

plt.show()
```

```python
# plot the difference between ESRI and ArcGIS Pro
ax = results.plot(kind='line',y=[ "ESRI Routing Service",  "ArcGIS Pro"],
                  linewidth=0.3,
                  alpha=0.9)
# plt.title("Comparison of Route Drive Durations")
plt.xlabel("Google Maps Drive Time (Minute)")
plt.ylabel("Drive Time (Minute)")
ax.set_xlim([0, 100])
ax.set_ylim([0, 150])
legend_lines = [mlines.Line2D([], [], color=c, linewidth=2) for c in plt.rcParams['axes.prop_cycle'].by_key()['color']]
legend_labels = [  "ESRI Routing Service",  "ArcGIS Pro"]
plt.legend(legend_lines, legend_labels,frameon=False,loc='upper left')

# plt.savefig("../reports/figures/Comparison_of_Route_Drive_Durations_with100.pdf")
# plt.show()
```
